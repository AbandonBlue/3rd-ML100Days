```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數 checked
- 實驗測試不同的初始值

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans    # kmeans estimator
%matplotlib inline
```

```python
estimators = [('k_means_8', KMeans(n_clusters=8)),
              ('k_means_5', KMeans(n_clusters=5)),
              ('k_means_3', KMeans(n_clusters=3)),
              ('k_means_3', KMeans(n_clusters=3))]
```

```python
fignum = 1
titles = ['8 clusters', '5 clusters', '3 clusters']
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4, 3))
    ax = Axes3D(fig)
    
    # fit data
    est.fit(X)
    
    # 怎麼來的啊?這啥?
    labels = est.labels_
    
    # 給進去的值為啥這樣?
    # 不同的特徵佔據不同的軸線
    ax.scatter(X[:, 0], X[:, 1], X[:, 2],
               c=labels.astype(np.float), edgecolor='k')

    ax.w_xaxis.set_ticklabels([])    # 設定text values of the tick labels 不太知道
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
#     ax.set_xlabel(X.labels[0])            # 設定軸的特徵名稱, 但我不知道...
#     ax.set_ylabel(X.labels[1])
#     ax.set_zlabel(X.labels[2])
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1
```
